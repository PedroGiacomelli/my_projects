```python
os.chdir("")# Your original repository path
init_ls=os.listdir()

path_o=r""# Your original repository path

def open_dirs(x,path_func):
    
    path_updater=lambda path_func: path_func+r"\{my_item}".format(my_item=item)
    
    if x:
        for item in x:
            
            if item.endswith((".txt",".csv",".pdf")):
                try:                             
                    csv_path=path_func.replace(r"Your original repository path","Your destination repository path")

                    cur_func = os.path.join(path_func,  item)

                    folder_path, filename = os.path.split(cur_func)

                    dst_folder = os.path.join(csv_path, os.path.splitext(filename)[0])  
                    
                    if not os.path.exists(dst_folder):
                        os.makedirs(dst_folder)
                        os.chmod(dst_folder, stat.S_IRWXU | stat.S_IRWXG | stat.S_IRWXO)

                    shutil.copy(cur_func, dst_folder)
                    print("OK - ",cur_func)
                    
                except:
                    if not os.listdir(dst_folder):
                        os.rmdir(dst_folder)
                    
                    fp,fn=os.path.split(dst_folder)
                    print("OK* - ",cur_func)
                    shutil.copy(cur_func, fp)                    
                
            elif item.endswith(".xlsx"):
                db_unwrap(path_func,item,True)
                
            elif item.endswith(".xls"):
                db_unwrap(path_func,item,False)
                
            else:
                
                os.chdir(path_updater(path_func))
                open_dirs(os.listdir(),path_updater(path_func))
    else:
        pass
```
